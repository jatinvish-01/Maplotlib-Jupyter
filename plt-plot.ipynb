```python
# plt.plot(x, y, color = "color name", linestyle="line_style", line_width=value in int, marker = "marker symbol", label = "label name" )

import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000,9000,3000,7800]
plt.plot(months, sales, color = "blue", linestyle="--", linewidth = 2, marker = "o", label = "2025 Sales Data" )

plt.xlabel("Number of Months")
plt.ylabel("Sales per Months")
plt.title("Monthly Sales Data Reports")
plt.legend(loc="lower right", fontsize=8)
plt.grid(color='gray', linestyle=":", linewidth = 1)
plt.xticks([1, 2, 3, 4], ['M1', 'M2', 'M3', 'M4'])# This will change the previces value of tyhe x axis


plt.show()
```
