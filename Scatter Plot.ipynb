### Scatter Plot

```python
#plt.scatter(x, y, color="color_name", marker="marker style", label="label_name")

import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [56,67,78,89,90,87,98,65]

plt.scatter(hours_studies, exam_scores, color="green", marker="o", label="Student Data")  
plt.xlabel("Hours Studies")
plt.ylabel("Exam Scores")
plt.title("Relationship between Study Times and Exam Scores")
plt.legend()
plt.grid(True)
plt.show()
```

##### For Two Groups

```python
import matplotlib.pyplot as plt

plt.scatter([1,2,3,4], [67,89,90,98], color="blue", label="Class A") #g1
plt.scatter([1,2,3,4], [56,76,87,66], color="orange", label="Class B") #g2

plt.xlabel("Hours Studies")
plt.ylabel("Exam Scores")
plt.title("Comparison of Two Classes")
plt.legend()
plt.grid(True)
plt.show()
```

##### Adding Color and Size

```python
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 26, 60, 65, 75, 67, 90]

colors = ["red" if score < 50 else "green" for score in exam_scores]
sizes = [score *2 for score in exam_scores]
plt.scatter(study_hours, exam_scores, c=colors, s=sizes)
plt.xlabel("Student Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
plt.grid(True)
plt.show()
```

##### Using a Colormap cmp=()

```python
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 26, 60, 65, 75, 67, 90]

# sizes = [score * 2 for score in exam_scores]
social_scores = [12, 34, 67, 678, 145, 56, 999, 445, 300] 

plt.scatter(study_hours, exam_scores, c=social_scores, cmap="viridis")

plt.colorbar(label="Social Score")
plt.xlabel("Student Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
plt.grid(True)
plt.show()
```

##### Adding Annotations

```python
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 85, 50, 26, 60, 45, 75, 67, 90]

# sizes = [score * 2 for score in exam_scores]
social_scores = [12, 34, 67, 678, 145, 56, 999, 445, 300] 

plt.scatter(study_hours, exam_scores, c=social_scores, cmap="viridis")

# Add labels
for i in range(len(study_hours)):
    plt.annotate(f'S {i+1}', (study_hours[i], exam_scores[i] - 3))

plt.colorbar(label="Social Score")
plt.xlabel("Student Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
plt.grid(True)
plt.show()
```
