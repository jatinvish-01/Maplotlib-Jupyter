```python
import matplotlib.pyplot as plt

# plt.subplot(nrows, ncols, index)

# Data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [5, 15, 20, 25, 40]

# First plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot(x, y1, color='blue')
plt.title('Plot 1')
plt.grid(True)

# Second plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.scatter(x, y2, color='green')
plt.title('Plot 2')
plt.grid(True)

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```

#### Object Oriented way to Plot a Subplot

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 3, figsize=(12, 4))

x = [1, 2, 3, 4, 5, 6]
y = [45, 67, 32, 98, 12, 44]

ax[0].plot(x, y, color="green")
ax[0].set_title("Line Chart")
ax[0].grid(True)

ax[1].bar(x, y)
ax[1].set_title("Bar Chart")

ax[2].scatter(x, y, color="orange", marker="o") 
ax[2].set_title("scatter Chart")
ax[2].grid(True)
 
fig.suptitle("Comparison of Line, Bar and Scatter chart") # To set the title of the Subplot

plt.tight_layout()

# To save the chart in this form of Extension
plt.savefig("Comparison_charts.png", dpi=300, bbox_inches="tight") 
plt.show()
```
