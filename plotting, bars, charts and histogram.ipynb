```python
import matplotlib.pyplot as plt # Bar charts
 
# Data
categories = ['A', 'B', 'C']
values = [10, 24, 36]

# Plot
plt.barh(categories, values)

# Optional
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

```python
import matplotlib.pyplot as plt # Pie charts

# Data
labels = ['Apple', 'Banana', 'Cherry']
sizes = [30, 45, 25]

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Optional
plt.title('Pie Chart')
plt.axis('equal')  # Makes it a circle

plt.show()
```

```python
import matplotlib.pyplot as plt # Histograms 

# Data
data = [22, 25, 27, 30, 32, 33, 35, 36, 40, 45, 50]

# Plot
plt.hist(data, bins=5, color='pink', edgecolor='black')

# Optional
plt.title('Histogram')
plt.xlabel('Value Range')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt

regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold', 'skyblue', 'lightgreen', 'coral'])
plt.title("Revenue Contributions by Region")
plt.show()
```

```python
#plt.hist(data, bins=numbers_of_bins, color='colorname', edgecolor='black')
import matplotlib.pyplot as plt

scores = [45,76,34,83,50,23,62,86,88,44,50,22]

plt.hist(scores, bins=6, color='purple', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score Distrubution of Students')
plt.show()
```

```python
#Scatter Plot
#plt.scatter(x, y, color="color_name", marker="marker style", label="label_name")

import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [56,67,78,89,90,87,98,65]

plt.scatter(hours_studies, exam_scores, color="green", marker="o", label="Student Data")  
plt.xlabel("Hours Studies")
plt.ylabel("Exam Scores")
plt.title("Relationship between Study Times and Exam Scores")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# For Two Groups
import matplotlib.pyplot as plt

plt.scatter([1,2,3,4], [67,89,90,98], color="blue", label="Class A") #g1
plt.scatter([1,2,3,4], [56,76,87,66], color="orange", label="Class B") #g2

plt.xlabel("Hours Studies")
plt.ylabel("Exam Scores")
plt.title("Comparison of Two Classes")
plt.legend()
plt.grid(True)
plt.show()
```
