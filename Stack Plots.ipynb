### Stack Plots in Matplotlib

```python
import matplotlib.pyplot as plt

activities = ["Studing", "Playing", "Watching TV", "Sleeping"]
time_spend = [4, 2, 3, 7]

colors = ["skyblue", "red", "pink", "yellow"]
plt.figure(figsize=(6, 6))
plt.pie(time_spend, labels=activities, colors=colors, startangle=90, autopct="%1.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt
 
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ["Studying", "Playing", "TV", "Sleeping"]
colors = ["skyblue", "red", "pink", "yellow"]

plt.figure(figsize=(6,6))
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels=labels, colors=colors, alpha=0.7)

plt.legend(loc="upper left")
plt.title("Week in the life of the student")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
plt.show()
```
