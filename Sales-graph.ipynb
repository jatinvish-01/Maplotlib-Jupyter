```python
import matplotlib.pyplot as plt

x = ['Mon', 'Tue', 'Wed', 'Thus', 'Fri']
y = [23, 45, 67, 12, 98]

plt.plot(x, y)


plt.title("Bakery sales in this week!!") 

plt.xlabel('Days of the week') 
plt.ylabel('Sales per day') 

plt.fill_between(x, y, color='skyblue', alpha=0.5, label='Sales Range') # coloring between the downside, and set the transprancy of it
plt.plot(x, y, color= 'yellow', label='Sales Line')

plt.legend()
plt.grid(True)
plt.show()
```
